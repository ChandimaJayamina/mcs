# Robin Karp algo

```python
def robin_karp(text, pattern, prime=101):
    n = len(text)
    m = len(pattern)
    d = 256  # Number of characters in the input alphabet
    p_hash = 0  # Hash value for pattern
    t_hash = 0  # Hash value for text
    h = 1  # The value of d^(m-1)
    result = []

    # The value of h would be "pow(d, m-1)%prime"
    for i in range(m-1):
        h = (h * d) % prime

    # Calculate the hash value of pattern and first window of text
    for i in range(m):
        p_hash = (d * p_hash + ord(pattern[i])) % prime
        t_hash = (d * t_hash + ord(text[i])) % prime

    # Slide the pattern over text one by one
    for i in range(n - m + 1):
        # Check the hash values of current window of text and pattern
        if p_hash == t_hash:
            # If the hash values match then only check for characters one by one
            if text[i:i+m] == pattern:
                result.append(i)

        # Calculate hash value for next window of text
        if i < n - m:
            t_hash = (d * (t_hash - ord(text[i]) * h) + ord(text[i + m])) % prime

            # We might get negative values of t_hash, converting it to positive
            if t_hash < 0:
                t_hash = t_hash + prime

    return result
```

```python
# Example usage
text = "abdabcaabc"
pattern = "abc"
positions = robin_karp(text, pattern)

if positions:
    print(f"Pattern found at positions: {positions}")
else:
    print("Pattern not found")
```

```python
def robin_karp(text, pattern, prime=101):
    collision_count = 0
    n = len(text)
    m = len(pattern)
    d = 256  # Number of characters in the input alphabet
    p_hash = 0  # Hash value for pattern
    t_hash = 0  # Hash value for text
    h = 1  # The value of d^(m-1)
    result = []

    # The value of h would be "pow(d, m-1)%prime"
    for i in range(m-1):
        h = (h * d) % prime

    # Calculate the hash value of pattern and first window of text
    for i in range(m):
        p_hash = (d * p_hash + ord(pattern[i])) % prime
        t_hash = (d * t_hash + ord(text[i])) % prime

    # Slide the pattern over text one by one
    for i in range(n - m + 1):
        # Check the hash values of current window of text and pattern
        if p_hash == t_hash:
            # If the hash values match then only check for characters one by one
            if text[i:i+m] == pattern:
                result.append(i)
            if text[i:i+m] != pattern:
                collision_count += 1

        # Calculate hash value for next window of text
        if i < n - m:
            t_hash = (d * (t_hash - ord(text[i]) * h) + ord(text[i + m])) % prime

            # We might get negative values of t_hash, converting it to positive
            if t_hash < 0:
                t_hash = t_hash + prime

    return result, collision_count
```

```python
def robin_karp(text, pattern, prime=101):
    n = len(text)
    m = len(pattern)
    d = 256  # Number of characters in the input alphabet
    p_hash = 0  # Hash value for pattern
    t_hash = 0  # Hash value for text
    h = 1  # The value of d^(m-1)
    result = []

    # The value of h would be "pow(d, m-1)%prime"
    for i in range(m-1):
        h = (h * d) % prime

    # Calculate the hash value of pattern and first window of text
    for i in range(m):
        p_hash = (d * p_hash + ord(pattern[i])) % prime
        t_hash = (d * t_hash + ord(text[i])) % prime

    # Slide the pattern over text one by one
    for i in range(n - m + 1):
        # Check the hash values of current window of text and pattern
        if p_hash == t_hash:
            # If the hash values match then only check for characters one by one
            if text[i:i+m] == pattern:
                result.append(i)

        # Calculate hash value for next window of text
        if i < n - m:
            t_hash = (d * (t_hash - ord(text[i]) * h) + ord(text[i + m])) % prime

            # We might get negative values of t_hash, converting it to positive
            if t_hash < 0:
                t_hash = t_hash + prime

    return result
```

## Collison Rate

```python
def robin_karp(text, pattern, prime=101):
    collision_count = 0
    n = len(text)
    m = len(pattern)
    d = 256  # Number of characters in the input alphabet
    p_hash = 0  # Hash value for pattern
    t_hash = 0  # Hash value for text
    h = 1  # The value of d^(m-1)
    result = []

    # The value of h would be "pow(d, m-1)%prime"
    for i in range(m-1):
        h = (h * d) % prime

    # Calculate the hash value of pattern and first window of text
    for i in range(m):
        p_hash = (d * p_hash + ord(pattern[i])) % prime
        t_hash = (d * t_hash + ord(text[i])) % prime

    # Slide the pattern over text one by one
    for i in range(n - m + 1):
        # Check the hash values of current window of text and pattern
        if p_hash == t_hash:
            # If the hash values match then only check for characters one by one
            if text[i:i+m] == pattern:
                result.append(i)
            if text[i:i+m] != pattern:
                collision_count += 1

        # Calculate hash value for next window of text
        if i < n - m:
            t_hash = (d * (t_hash - ord(text[i]) * h) + ord(text[i + m])) % prime

            # We might get negative values of t_hash, converting it to positive
            if t_hash < 0:
                t_hash = t_hash + prime

    return result, collision_count
```

```python
file_path = './data/dna_sequence_25000.txt'  # Replace with your file path
text = read_text_from_file(file_path)
pattern = "CTCAGTG"  # Define the pattern you want to search for

# Your function call here
positions, collision_count101 = robin_karp(text, pattern, 101)

if positions:
    print(f"Pattern found at positions: {positions}")
else:
    print("Pattern not found")
# After running the algorithm, check collision_count
print(f"Number of hash collisions: {collision_count101}")
```
